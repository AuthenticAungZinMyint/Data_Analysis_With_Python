```python
#drawing on images
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

image_shape = (600, 600, 3)
#create black image
black_image = np.zeros(shape=image_shape, dtype=np.int16)
plt.imshow(black_image)
```

```python
#create white image
image_shape = (600, 600, 3)
white_image = np.zeros(shape=image_shape, dtype=np.int16)

#every pixel of the image to 255
white_image.fill(255)
plt.imshow(white_image)
```

```python
#draw a line in black image
line = cv2.line(black_image, (599,0), (0,599), (0,255,0),4)
plt.imshow(line)
```

```python
#draw line on white image
line = cv2.line(white_image, (0,599), (599,0), (0,0,255), 4)
plt.imshow(line)
```

```python
#create circle a white image
img_shape = (600, 600, 3)
white_image = np.zeros(shape= img_shape, dtype=np.int16)

white_image.fill(255)

#draw a red circle on white image
circle=cv2.circle(white_image, (300, 300), 100, (255,0,0), 6)

plt.imshow(circle)
```

```python
#draw rectangle on black image
img_shape = (600, 600, 3)
black_image = np.zeros(shape = img_shape, dtype = np.int16)

#draw green rectangle
rectangle = cv2.rectangle(black_image, (200,200), (400,500), (0,255,0), 5)
plt.imshow(rectangle)
```

```python
#writing text on images

img_shape = (600, 800, 3)
black_image = np.zeros(shape=img_shape, dtype=np.int16)

# putText() function will take the following arguments: image, text, coordinates of the bottom-left corner, font, fontScale, color, thickness, and linetype.
text = cv2.putText(black_image, 'Thanksgiving', (10,500), cv2.FONT_HERSHEY_SIMPLEX,
                  3, (255,0,0), 2, cv2.LINE_AA)
plt.imshow(text)
```

```python
#changing brightness
# The value of pixel intensity can help us to find a brighter image. For example, if two pixels have the intensity values 110 and 230, then the latter one is brighter. 

import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

image = cv2.imread('image/nature.jpeg')
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image)
```

```python
#change brightness
#set weightage for alpha and betaboth the matrix
alpha_ = 1
beta_ = 50

#add weight to the original image to change the brightness
image_change = cv2.addWeighted(rgb_image, alpha_,
                              np.zeros(image.shape, image.dtype), 0, beta_)
plt.imshow(image_change)
```

Gaussian Blur
  In this blurring, the Gaussian kernel is used instead of a box filter. GaussianBlur() will take the image and kernel size. The kernel size will be a tuple of the width and height. Both width and height must be a positive and odd number:    

```python
#Gaussian kernel (Gaussian Blur)
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

image = cv2.imread('image/tajmahal.jpg')
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image_blur = cv2.GaussianBlur(rgb_image, (7,7), 0)

plt.imshow(image_blur)
```

Median blur
    medianBlur() will take image and kernel size as an argument. It is recommended that the kernel size should be an odd number and greater than 1, for example, 3, 5, 7, 9, 11, and so on: 

```python
#median blur

image_blur_1 = cv2.medianBlur(rgb_image, 11)
plt.imshow(image_blur_1)
```

Face detection
    Face detection is a kind of classification problem. We can classify images into two classes, face or not face. We need lots of images to train such a model for classification. Thankfully, OpenCV offers pre-trained models such as the Haar Feature-Based Cascade Classifier and the Local Binary Pattern (LBP) classifier, trained on thousands of images. In our example, we will use Haar feature extraction to detect a face. 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

image = cv2.imread('image/messi.png')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(image_gray)
```

```python
#load Haar Cascade face classifier file
haar_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

```

```python
#get the faces coordinate for all the faces in the image
faces_coordinates = haar_cascade.detectMultiScale(image_gray,
                                                scaleFactor = 1.3,
                                                minNeighbors = 7);
#draw rectangle on detected face
for (p, q, r, s) in faces_coordinates:
    cv2.rectangle(image, (p, q), (p + r, q + s), (255, 255, 0), 2)

#convert bgr to rgb
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
```

```python
#multi face detection
image = cv2.imread('image/barcelona.jpeg')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

haar_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

face_coordinates = haar_cascade.detectMultiScale(image_gray, scaleFactor = 1.3, minNeighbors = 5);

for (p, q, r, s) in face_coordinates:
    cv2.rectangle(image, (p, q), (p+r, q+s), (255,255,0), 2)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```
